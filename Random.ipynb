```python
# prompt: generate the binomial distribution of random variables

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters of the binomial distribution
n = 10  # Number of trials
p = 0.3  # Probability of success

# Generate random variables from the binomial distribution
random_variables = binom.rvs(n=n, p=p, size=1000)

# Create a histogram of the random variables
plt.hist(random_variables, bins=range(n + 2), align='left', rwidth=0.8)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n = {}, p = {})'.format(n, p))
plt.xticks(range(n + 1))
plt.show()
```

```python
# prompt: by using above data generate the differnt types of graphs using highend frame work

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom
import seaborn as sns
import plotly.express as px

# Parameters of the binomial distribution (same as before)
n = 10  # Number of trials
p = 0.3  # Probability of success

# Generate random variables (same as before)
random_variables = binom.rvs(n=n, p=p, size=1000)

# 1. Histogram (using matplotlib, as in original code)
plt.hist(random_variables, bins=range(n + 2), align='left', rwidth=0.8)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n = {}, p = {}) - Histogram (Matplotlib)'.format(n, p))
plt.xticks(range(n + 1))
plt.show()


# 2. Histogram (using seaborn)
sns.histplot(random_variables, bins=range(n + 2), kde=False)  # kde=False to remove density curve
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n = {}, p = {}) - Histogram (Seaborn)'.format(n, p))
plt.xticks(range(n + 1))
plt.show()

# 3. Bar plot (using matplotlib)
unique, counts = np.unique(random_variables, return_counts=True)
plt.bar(unique, counts)
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.title('Binomial Distribution (n = {}, p = {}) - Bar Plot (Matplotlib)'.format(n, p))
plt.xticks(range(n+1))
plt.show()


# 4. Interactive Histogram (using plotly)
fig = px.histogram(x=random_variables, nbins=n+1,
                   labels={'x':'Number of Successes', 'y':'Frequency'},
                   title='Binomial Distribution (n = {}, p = {}) - Interactive Histogram (Plotly)'.format(n, p))
fig.show()


# 5. Probability Mass Function (using matplotlib) - Theoretical distribution
x = np.arange(0, n + 1)
probabilities = binom.pmf(x, n, p)
plt.bar(x, probabilities)
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.title('Binomial Distribution (n = {}, p = {}) - Probability Mass Function'.format(n, p))
plt.show()
```
